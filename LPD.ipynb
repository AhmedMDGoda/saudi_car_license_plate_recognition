```python
from ultralytics import YOLO

# Load a YOLOv8 model (you can replace 'yolov8n.pt' with 'yolov8s.pt', 'yolov8m.pt', etc.)
model = YOLO('yolov8n.pt')
model.train(data='/kaggle/input/saudi-carplate/saudi_ALPR.v2i.yolov8/data.yaml', epochs=50, imgsz=640)
```

```python
import pandas as pd
df = pd.read_csv('/kaggle/working/runs/detect/train2/results.csv')
print(df)
```

```python
print('mAP = ',model.metrics.results_dict['metrics/mAP50(B)'])
```

```python
model2 = YOLO('/kaggle/working/runs/detect/train2/weights/best.pt')
```
