```python
model2 = YOLO('/kaggle/working/runs/detect/train/weights/best.pt')
predictions = model2('/kaggle/input/reco-dataset/RECO_DS/test/images/-1_png_jpg.rf.323dca16b4b75a47c99e45b4e45e0b92.jpg')[0]
boxes_with_labels = []
for result in predictions.boxes.data.tolist():
    x1, y1, x2, y2, confidence, class_id = result
    class_name = predictions.names[class_id]
    boxes_with_labels.append((x1, class_name))
sorted_boxes = sorted(boxes_with_labels)
text = ''
for box in sorted_boxes:
    text += box[1]
print(text)
```
